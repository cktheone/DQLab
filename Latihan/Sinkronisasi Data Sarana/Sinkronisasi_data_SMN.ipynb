```python
# replace missing value(null value) menjadi string "NA" untuk pendefinisian nilai
merge_df['Nomor'] = merge_df['Nomor'].fillna("NA")
merge_df['Area'].fillna("NA", inplace=True)
merge_df['kode_daop'].fillna(0, inplace=True)
```

```python
merge_df.dtypes
```

```python
#fungsi untuk mengupdate kolom status berdasrkan data rujukan
# AWAS, jika ada nomor yang mengalami perubahan dan belum terupdate maka status akan otomatis berubah ke Tidak Aktiv
# perlu pengecualian untuk data smn yg tidak tercatat seperti KFW, SN01401, SK11901
def func_status(row):
    if row['kode_sarana'] == "SN01401" or row['kode_sarana'] == "SK11901" :
        status = "Aktiv"
    elif row['id_jenis'] == 2 :
        status = "Aktiv"
    elif row['Nomor'] == "NA" :
        status = "Tidak Aktiv"
    else:
        status = "Aktiv"
    
    return status
```

```python
#cek isi kolom status
merge_df['status'].value_counts(dropna=False)
```

```python
# mengupdate status sesuai dg data rujukan
merge_df['status']= merge_df.apply(func_status, axis=1)
merge_df['status'].value_counts(dropna=False)
```

```python
#fungsi untuk merubah nilai pada kolom DAOP yg NA diganti dengan isi dari kolom nama_daop
def ubah_daop(x,y):
    if  x == 0 :
        daop = y 
    else :
        daop = x
        
    return daop 
```
