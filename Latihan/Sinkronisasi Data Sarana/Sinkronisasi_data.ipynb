```python
# replace missing value(null value) menjadi string "NA" untuk pendefinisian nilai
merge_df['NOMOR_IDENTITAS'] = merge_df['NOMOR_IDENTITAS'].fillna("NA")
merge_df['DAOP'].fillna("NA", inplace=True)
merge_df['JENIS'].fillna("NA", inplace=True)
```

```python
#fungsi untuk mengupdate kolom status berdasrkan data rujukan
# AWAS, jika ada nomor yang mengalami perubahan dan belum terupdate maka status akan otomatis berubah ke Tidak Aktiv
def fx(x):
    if x == "NA":
        return "Tidak Aktiv"
    else:
        return "Aktiv"
```

```python
#cek isi kolom JENIS
merge_df['jenis_sarana'].value_counts()
```

```python
#mengubah string ke upper pada kolom jenis_sarana agar sama dengan yg di kolom JENIS
merge_df['jenis_sarana'] = merge_df['jenis_sarana'].apply(lambda x: x.upper())
merge_df['jenis_sarana'].value_counts()
```

```python
#fungsi untuk merubah jenis sarana bila ada perubahan/modifikasi ex: bagasi ke SN 
#dg catatan masih menggunakan no lama belum ada penetapanya tp secara fungsi sudah diganti
def ubah_jenis(x,y):
    jenis_sarana = ""
    if x == y :
        jenis_sarana = x
    elif y == 'NA' :
        jenis_sarana = x
    
    else :
        jenis_sarana = y
        
    return jenis_sarana 
```

```python
# mengupdate status sesuai dg data rujukan
merge_df['status']= merge_df['NOMOR_IDENTITAS'].apply(lambda x : fx(x))
```

```python
# mengupdate perubahan jenis sarana (bilamana diperlukan)
merge_df['jenis_sarana']= list(map(lambda x, y : ubah_jenis(x,y), merge_df['jenis_sarana'], merge_df['JENIS']))
merge_df
```

```python
#fungsi digunakan untuk mengubah jenis sarana ke kode_sarana sesuai database (sesuaikan apabila ada perubahan)
def kode_sarana(jenis_sarana):
    if jenis_sarana == 'LOKOMOTIF' :
        jenis_sarana = 1
    elif jenis_sarana == 'KRL' :
        jenis_sarana = 2
    elif jenis_sarana == 'LRT' :
        jenis_sarana = 2
    elif jenis_sarana == 'KRD' :
        jenis_sarana = 3
    elif jenis_sarana == 'GERBONG' :
        jenis_sarana = 4 
    elif jenis_sarana == 'PERALATAN KHUSUS' :
        jenis_sarana = 9 
    elif jenis_sarana == 'KERETA' :
        jenis_sarana = 11
    elif jenis_sarana == 'RAILBUS' :
        jenis_sarana = 19
    else :
        jenis_sarana = 'undefined'
        
    return jenis_sarana 
```

```python
# ubah nama jenis ke kode jenis
merge_df['jenis_sarana'] = merge_df['jenis_sarana'].apply(lambda x: kode_sarana(x))
```

```python
#cek hasilnya apakah ada yg undefined/null??
merge_df['jenis_sarana'].value_counts(dropna=False)
```

```python
#fungsi untuk merubah nilai pada kolom DAOP yg NA diganti dengan isi dari kolom nama_daop
def ubah_daop(x,y):
    daop = ""
    if  x == 'NA' :
        daop = y 
    else :
        daop = x
        
    return daop 
```

```python
#ubah na pada kolom daop dengan memanggil fungsi yang telah dibuat cek hasilnya
merge_df['DAOP']= list(map(lambda x, y : ubah_daop(x,y), merge_df['DAOP'], merge_df['nama_daop']))
merge_df['DAOP'].value_counts(dropna=False)
```

```python
# fungsi untuk ubah nama daop pada kolom DAOP menjadi kode daop(sesuaikan dengan database) 
def kode_daop(daop):
    if daop == 'DAOP 1' or daop == 'DAOP 1 JAKARTA' :
        daop = 7
    elif daop == 'DAOP 2' or daop == 'DAOP 2 BANDUNG' :
        daop = 8
    elif daop == 'DAOP 3' or daop == 'DAOP 3 CIREBON':
        daop = 9
    elif daop == 'DAOP 4' or daop == 'DAOP 4 SEMARANG' :
        daop = 10
    elif daop == 'DAOP 5' or daop == 'DAOP 5 PURWOKERTO' :
        daop = 11 
    elif daop == 'DAOP 6' or daop == 'DAOP 6 YOGYAKARTA':
        daop = 12 
    elif daop == 'DAOP 7' or daop == 'DAOP 7 MADIUN' :
        daop = 13
    elif daop == 'DAOP 8' or daop == 'DAOP 8 SURABAYA' :
        daop = 14
    elif daop == 'DAOP 9'  or daop == 'DAOP 9 JEMBER':
        daop = 15 
    elif daop == 'DIVRE 1' or daop == 'DIVRE I SUMATERA UTARA' :
        daop = 16 
    elif daop == 'DIVRE 2' or daop ==  'DIVRE II SUMATERA BARAT':
        daop = 17
    elif daop == 'DIVRE 3' or daop ==  'DIVRE III SUMATERA SELATAN':
        daop = 18
    elif daop == 'DIVRE 4' or daop == 'DIVRE IV LAMPUNG' :
        daop = 26
    elif daop == 'BY LT' :
        daop = 18
    elif daop == 'BY MRI' :
        daop = 26
    elif daop == 'PT KAI COMMUTER JABOTABEK' :
        daop = 24
    else :
        daop = 'undefined'
        
    return daop 
```

```python
#panggil fungsi untuk mengubah nama menjadi kode
merge_df['DAOP'] = merge_df['DAOP'].apply(lambda x: kode_daop(x))
```

```python
#cek hasilnya apakah masih ada yg undefined, lakukan pengecekan kemali jika masih ada
merge_df['DAOP'].value_counts()
```
