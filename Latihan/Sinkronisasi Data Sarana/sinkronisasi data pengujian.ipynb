```python
import pandas as pd
```

```python
# ambil data yang mau di kroscek status sertifikasinya (data kai)
df_kai_baru = pd.read_csv('../data_source/datsar/data_kai_after_okt.csv', index_col=0)
df_kai_baru

```

```python
#ambil data master sarana dari tanda lulus uji
colnames=['id', 'kode_sarana', 'kode_sarana_lama', 'id_kartu', 'id_jenis', 'id_sub_jenis', 'seri', 'tahun_dinas', 'id_daops', 'id_depo', 'status', 'negara_asal', 'tgl_uji_pertama', 'tgl_masa_uji', 'tgl_masa_perawatan', 'pabrikan', 'pemilik', 'alamat', 'keterangan' ] 
df_sarana = pd.read_csv('../data_source/datsar/data_sarana_okt.csv', names=colnames)
df_sarana.head()
```

```python
merge_df = pd.merge(left=df_kai_baru, right=df_sarana, how='left', left_on='NOMOR_IDENTITAS', right_on='kode_sarana')
merge_df.shape
```

```python
merge_df = merge_df[['NOMOR_IDENTITAS', 'DAOP', 'DEPO', 'JENIS', 'tgl_masa_uji']]
merge_df
```

```python
merge_df.dtypes
```

```python
merge_df['tgl_masa_uji'] = merge_df['tgl_masa_uji'].str.replace('0000-00-00', '2007-01-01')
```

```python
merge_df['tgl_masa_uji']= pd.to_datetime(merge_df['tgl_masa_uji'])
```

```python
merge_df.dtypes
```

```python
merge_df
```

```python
import datetime
now = datetime.date.today()
print(now)
```

```python
def status_uji(x):
    
    
    if x > now:
        return "Masih Berlaku"
    else:
        return "Kadaluarsa"
```

```python
merge_df['STATUS_UJI'] = merge_df['tgl_masa_uji'].apply(lambda x : status_uji(x))
```
