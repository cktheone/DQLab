Normalisasi tabel data sarana (data dr PT KAI)
- digunakan untuk menyusun dataset sarana milik PT KAI dari file excel yang tadinya di bedakan per sheet sesuai jenisnya di gabung menjadi satu agar dapat diolah lebih lanjut untuk jadi bahan benchmark sinkronisasi data (relate to Sinkronisasi_data.ipynb)
- output dataset memiliki kolom (NOMOR_IDENTITAS, DAOP, DEPO, JENIS, TAHUN, NAMA_DAOP2, NAMA_DEPO, KODE_DEPO), ekspor ke csv file untuk pengolahan lebih lanjut
- untuk data gerbong pastikan format nomor sudah sesuai, pastikan di file excl untuk merubahnya terutama untuk yg 3 digit no urut (001) harus dirubah ke 01 (GD4214071-GD4214080, GB5012001-099, GB5013001-099, GB5014001-099 dll), hilangkan header pada file excel di tiap jenis sarana, format file xlsx
- tandain nomor yang terjadi perubahan tapi belum ada penetapan sk ny.
- terdapat fitur untuk normalisasi nilai Daop agar konsisten
- sudah ditambahkan kode_depo sesuai database bangwas (depo peralatan khusus belum)

```python
def func_depo(col):
    if col['JENIS'] == 'LOKOMOTIF' or col['JENIS'] == 'KRD' :
        depo_baru = "DEPO LOKOMOTIF " + col['DEPO']
    elif col['JENIS'] == 'KERETA' :
        depo_baru = "DEPO KERETA " + col['DEPO']
    elif col['JENIS'] == 'GERBONG' :
        depo_baru = "DEPO GERBONG " + col['DEPO']
    elif col['JENIS'] == 'KRL' :
        depo_baru = "DEPO KRL " + col['DEPO']
    else :
        depo_baru = col['DEPO']
    
    return depo_baru
```
