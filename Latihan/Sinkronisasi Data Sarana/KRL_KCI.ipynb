```python
import pandas as pd
colnames=['id', 'kode_sarana', 'kode_sarana_lama', 'id_kartu', 'id_jenis', 'id_sub_jenis', 'seri', 'tahun_dinas', 'id_daops', 'id_depo', 'status', 'negara_asal', 'tgl_uji_pertama', 'tgl_masa_uji', 'tgl_masa_perawatan', 'pabrikan', 'id_pemilik', 'created_at', 'last_update' ] 
datsar = pd.read_csv('../data_source/tblsarana.csv', names=colnames)

datsar.head()
```

```python
datsar_kci = datsar.loc[(datsar['id_jenis'] == 2) & (datsar['id_pemilik'] == 3)]
```

```python
jumlah_krl = datsar_kci.groupby(['status'])['kode_sarana'].count().reset_index()
print(jumlah_krl)
```

```python
import matplotlib.pyplot as plt
plt.clf()
datsar_kci.groupby(['status'])['kode_sarana'].count().plot(kind='bar', color='green')
plt.title('Status Armada KRL Milik PT KCI 2021',loc='center',pad=30, fontsize=15, color='blue')
plt.xlabel('Status', fontsize = 15)
plt.ylabel('Jumlah',fontsize = 15)
plt.ylim(ymin=0)
plt.xticks(rotation=0)
plt.show()
```

```python
def tahun_pembuatan(tahun) :
    tahun2 = re.findall(r'(\d{2})', tahun)
    tahun = tahun2[1]
    tahun = '19'+tahun
    
    
    return tahun
```

```python
datsar_kci['tahun_pembuatan'] = datsar_kci['kode_sarana'].apply(lambda x: tahun_pembuatan(x)).astype('int64')
datsar_kci.head()
```

```python
datsar_kci.dtypes
```

```python
def range_usia(umur) :
    if umur > 40 :
        usia = "40++"
    elif umur > 30 & umur <= 40 :
        usia = "31-40"
    elif umur > 20 & umur <= 30 :
        usia = "21-30"
    else :
        usia = "< 20"
    return usia
```

```python
#datsar_kci['usia'] = (2020 - datsar_kci['tahun_pembuatan']).apply(lambda x: "30++" if x > 30 else "< 30")
datsar_kci['usia'] = 2020 - datsar_kci['tahun_pembuatan']
```

```python
datsar_kci['range_usia'] = datsar_kci['usia'].apply(lambda x : range_usia(x))
```
