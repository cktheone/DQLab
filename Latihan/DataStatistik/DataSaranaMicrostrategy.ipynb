# fungsi untuk mengganti id jenis ke nama jenis sarana

```python
def transform_data(jenis_sarana):
    if jenis_sarana == 1 :
        jenis_sarana = 'Lokomotif'
    elif jenis_sarana == 2 :
        jenis_sarana = 'KRL'
    elif jenis_sarana == 3 :
        jenis_sarana = 'KRD'
    elif jenis_sarana == 4 :
        jenis_sarana = 'Gerbong'
    elif jenis_sarana == 9 :
        jenis_sarana = 'Peralatan Khusus'
    elif jenis_sarana == 11 :
        jenis_sarana = 'Kereta'
    elif jenis_sarana == 19 :
        jenis_sarana = 'RailBus'
    else :
        jenis_sara = 'undefined'
        
    return jenis_sarana 
```

```python
join_df2['id_jenis'] = join_df2['id_jenis'].apply(lambda x: transform_data(x))
join_df2 = join_df2.rename(columns={"id_jenis": "jenis_sarana"})
print(join_df2.head())
print(join_df2.tail())
```

# fungsi untuk membuat kategori jawa/sumatera berdasarkan nama_Daop (lihat hasil sebelumnya)

```python
import re
def func(col) :
    if re.match(r'\w+', col['nama_daop'])[0] == "DIVRE" :
        area = "SUMATERA"
    elif col['nama_daop'] == "BTP Sumut" or col['nama_daop'] == "BTP Sumbar" or col['nama_daop'] == "BTP Sumsel" :
        area = "SUMATERA"
    elif col['nama_daop'] == "BTP Sulsel" :
        area = "SULAWESI"
    else :
        area = "JAWA"
    return area
```

```python
data_sarana['lokasi'] = data_sarana.apply(func, axis=1)
```

```python
import re
def tahun(tahun) :
    tahun2 = re.findall(r'(0\d{2})', tahun)
    tahun = tahun2[0]
    return tahun

def ambiltahun(col) :
    if col['jenis_sarana'] == "Kereta" :
        x = col['kode_sarana']
        tahun2 = re.findall(r'(0\d{2})', x)
        tahun = tahun2[0]
    elif col['jenis_sarana']=='Lokomotif' :
        tahun = col['kode_sarana'][5:7]
    elif col['jenis_sarana']=='Gerbong' :
        tahun = col['kode_sarana'][4:6]
    else :
        tahun = col['kode_sarana'][3:5]
        
    return  tahun
```

```python
data_sarana['tahun'] = data_sarana.apply(ambiltahun, axis=1)
```

```python
data_sarana[data_sarana['jenis_sarana'] == 'Gerbong']
```

```python
data_sarana['tahun'] = data_sarana['tahun'].astype('int64')
```

```python
#fungsi untuk mengubah 2 digit tahun menjadi 4 digit
def func(row):
	if row['tahun'] <= 21:
		val = 2000 + row['tahun']
	else:
		val = 1900 + row['tahun']
	return val
# Tambahkan kolom baru
data_sarana['tahun_pembuatan'] = data_sarana.apply(func, axis=1)
```

```python
import datetime
now = datetime.datetime.now().year
print(now)
```

```python
#digunakan untuk menghitung umur rumus tahun sekarang - tahun pembuatan/dinas
data_sarana['umur'] = data_sarana['tahun_pembuatan'].apply(lambda x : datetime.datetime.now().year - x)
```

```python
data_sarana
```

```python
#membuat kategori range usia
def func_rangeusia(row):
	if row['umur'] <= 10:
		val = '<=10 th'
	elif (row['umur'] > 10 and row['umur'] <= 20):
		val ='11-20 th'
	elif (row['umur'] > 20 and row['umur'] <= 30):
		val ='21-30 th'
	else:
		val ='>30 th'
	return val
# Tambahkan kolom baru
data_sarana['range_usia'] = data_sarana.apply(func_rangeusia, axis=1)
```
