```python
import re
def daop(col) :
    if col['Lokasi Penempatan'] == "Gudang Peti Kemas Gedebage" or col['Lokasi Penempatan'] == "Gudang Prasarana Jatibarang" or col['Lokasi Penempatan'] == "Bandung":
        daop = 'BTP Wil. Jabar'
    elif col['Lokasi Penempatan'] == "Gudang Prasarana Pekalongan" or col['Lokasi Penempatan'] == "Gudang Prasarana Kroya" or col['Lokasi Penempatan'] == "Solo":
        daop = 'BTP Wil. Jateng'
    elif col['Lokasi Penempatan'] == "Gudang Prasarana Bangil" or col['Lokasi Penempatan'] == "PT INKA Madiun" :
        daop = 'BTP Wil. Jatim'
    elif col['Lokasi Penempatan'] == "Gudang Prasarana Parungpanjang" or col['Lokasi Penempatan'] == "Depo Lokomotif Tanahabang" or col['Lokasi Penempatan'] == "Depok" :
        daop = 'BTP Wil. JakBan'
    elif col['Lokasi Penempatan'] == "Depo Cipinang" or col['Lokasi Penempatan'] == "Depo Kereta Cipinang" :
        daop = 'BTP Wil. JakBan'
    elif col['Lokasi Penempatan'] == "Sumatera Utara" :
        daop = 'BTP Wil. Sumut'
    elif col['Lokasi Penempatan'] == "Sumatera Selatan" or col['Lokasi Penempatan'] == "Gudang Prasarana Payakabung" or col['Lokasi Penempatan'] == "Lampung":
        daop = 'BTP Wil. Sumsel'
    elif col['Lokasi Penempatan'] == "Sulawesi Selatan" :
        daop = 'BTP Wil. Sulsel'
    else :
        daop = 'Workshop Balai Perawatan'
        
    return  daop
```

```python
smn['daop'] = smn.apply(daop, axis=1)
```

```python
smn['daop'].value_counts(dropna = False)
```

```python
smn['daop'] = smn['daop'].str.replace('BTP Wil. ', 'BTP ')
```

```python
smn['daop'] = smn['daop'].str.replace('Workshop Balai Perawatan', 'BALAI PERAWATAN')
```
