```python
depo['Jenis_2'] = depo['Nama_Daop'].str.extract(r"(?:\s)(\w+)", expand=False)
```

```python
depo['lokasi'] = depo['Nama_Daop'].str.extract(r"(\w+)$", expand=False)
```

```python
def func(col) :
    if re.match(r'\w+', col['Daop'])[0] == "DAOP" :
        area = "JAWA"
    else :
        area = "SUMATERA"
    return area
        
    
```

```python

depo['lokasi'] = depo.apply(func, axis=1)
```

```python
depo['status_regex'] = depo['Daop'].str.contains(r'DAOP')
```
