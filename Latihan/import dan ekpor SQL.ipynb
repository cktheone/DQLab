Import Data From Mysql dan Ekspor ke CSV dalam direktori lokal

ada banyak cara/library yg dapat digunakan untuk mengimport file sql ke dataframe pandas
salah satunya library mysql.connector dan sqlalchemy

```python
#cara pertama ambil tabel sql kemudian ubah menjadi dataframe
import pandas as pd
import mysql.connector as sql

config = {
 'user': 'root',
 'password': '',
 'host': '127.0.0.1',
 'database': 'datsar',
 'raise_on_warnings': True,
 'use_pure':True 
}

db_connection = sql.connect(**config)
db_cursor = db_connection.cursor()
db_cursor.execute('SELECT * FROM tblsarana')

table_rows = db_cursor.fetchall()
colnames=['id', 'kode_sarana', 'kode_sarana_lama', 'id_kartu', 'id_jenis', 'id_sub_jenis', 'seri', 'tahun_dinas', 'id_daops', 'id_depo', 'status', 'negara_asal', 'tgl_uji_pertama', 'tgl_masa_uji', 'tgl_masa_perawatan', 'pabrikan', 'id_pemilik', 'created_at', 'last_update' ] 

df = pd.DataFrame(table_rows,  columns=colnames)
df.head()
```

```python
#cara kedua menggunakan pandas read_sql langsung
```

```python
try:
    mydb = sql.connect(**config)
    query = "Select * from tblsarana;"
    df2 = pd.read_sql(query,mydb)
    mydb.close() #close the connection
except Exception as e:
    mydb.close()
    print(str(e))
```

EKSPOR DATA TO MYSQL SERVER
pastikan library pymysql dan sqlalchemy sudah terinstall, jika belum install terlebih dahulu (cek dg conda list/pip list)

```python
#ambil data yang akan diupload ke dalam server mysql
depo = pd.read_csv("data_source/depo_after.csv", index_col=0)
depo.head()
```

```python
import pymysql
pymysql.install_as_MySQLdb()
import MySQLdb
from sqlalchemy import create_engine
#engine = create_engine('mysql://scott:tiger@localhost:5432/mydatabase') - setup koneksi lengkap dg passoword dan port akses mysql
engine = create_engine("mysql://root@localhost/datsar")

```

Aktivkan Kode di bawah jika akan upload data ke database mysql server, sesuaiakan dengan koneksi dan databenya

try :
    depo.to_sql(name='tbldepos_update', con=engine, index=False, if_exists='replace')
    print('Sucessfully written to Database!!!')

except Exception as e :
    print(e)

```python
#cara ketiga import sql menggunakan library sqlalchemy -perhatikan line 10 untuk setting engine/koneksinya
with engine.connect() as conn, conn.begin():
    data = pd.read_sql_table("tbldepos_update", conn)
```

```python
#cara keempat import sql menggunakan library sqlalchemy dengan query SQL
slicing = pd.read_sql_query("SELECT * FROM tbldepos_update WHERE Jenis = \"Depo Sarana Penggerak\"", engine)
```
