```python
import pandas as pd
df_participant = pd.read_csv('dqthon-participants.csv', sep=";")
```

```python
df_participant
```

```python
df_participant['postal_code'] = df_participant['address'].str.extract(r'(\d+)$') 
#asumsi no kodepos pada kalimat terakhir dan berupa angka pada kolom adresss
```

```python
df_participant
```

```python
df_participant['city'] = df_participant['address'].str.extract(r'(?:\d+)\n+(\w+)') 
#asumsi nama kota setelah no jalan/no rt/rw dan enter pada kolom adress sampai dengan spasi selanjutnya
```

```python
df_participant
```

```python
df_participant['github_profile'] = 'https://github.com/' + df_participant['first_name'].str.lower() + df_participant['last_name'].str.lower()
```

```python
df_participant
```

```python
#replace +62/atau62 menjadi 0
df_participant['cleaned_phone_number'] = df_participant['phone_number'].str.replace(r'^(\+62|62)', '0')
#replace ()/- menjadi ''
df_participant['cleaned_phone_number'] = df_participant['cleaned_phone_number'].str.replace(r'[()-]', '')
#menghapus spasi
df_participant['cleaned_phone_number'] = df_participant['cleaned_phone_number'].str.replace(r'\s+', '')
```

```python
df_participant
```

```python
def func(col):
    abbrev_name = "%s%s"%(col['first_name'][0].upper(),col['last_name'][0].upper()) #Singkatan dari Nama Depan dan Nama Belakang dengan mengambil huruf pertama
    country = col['country']
    abbrev_institute = '%s'%(''.join(list(map(lambda word: word[0], col['institute'].split())))) #Singkatan dari value di kolom institute
    return "%s-%s-%s"%(abbrev_name,country,abbrev_institute)

df_participant['team_name'] = df_participant.apply(func, axis=1)
```
